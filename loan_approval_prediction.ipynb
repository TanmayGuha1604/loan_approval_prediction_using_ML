```python
import pandas as pd

data = pd.read_csv('/content/loan_approval_dataset.csv')
data
```

```python
data.drop(columns=['loan_id'], inplace=True)
data
```

```python
data.columns
```

```python
data.columns=data.columns.str.strip()
data.columns
```

```python
data
```

```python
# let us combine all the assets columns into one
```

```python
data.columns
```

```python
data['Assets'] = data.residential_assets_value + data.commercial_assets_value + data.luxury_assets_value + data.bank_asset_value
```

```python
data
```

```python
data.drop(columns=['residential_assets_value','commercial_assets_value','luxury_assets_value','bank_asset_value'], inplace=True)
```

```python
data
```

```python
# To improve the performance of the machine learning model we will convert the alphabetical or string values into the numeric values. (We have 3 string valued columns: Education, self_employed and loan_status)
```

```python
# Before the above process check that there are any duplicate values or not
```

```python
data.isnull().sum()
```

```python
data.education.unique()
```

```python
# Here are some extra spaces in these columns, so we have to remove them.
```

```python
def clean_data(st):
  st = st.strip()
  return st
```

```python
clean_data(' Graduate')
```

```python
data.education = data.education.apply(clean_data)
```

```python
data.education.unique()
```

```python
# Now we will replace this Graduate and non-graduate with numeric values (1=Graduated and 0=Not Graduated)
```

```python
data['education'] = data['education'].replace(['Graduate','Not Graduate'],[1,0])
```

```python
data
```

```python
# The same process will be applied for self employed and and loan status
```

```python
data.self_employed.unique()
```

```python
data.self_employed = data.self_employed.replace(['No', 'Yes'],[0,1])
```

```python
data
```

```python
data.loan_status.unique()
```

```python
data.loan_status = data.loan_status.apply(clean_data)
```

```python
data.loan_status.unique()
```

```python
data.loan_status = data.loan_status.replace(['Approved','Rejected'],[1,0])
```

```python
data
```

```python
# Now we will split this data set into training dataset and test dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
input_data = data.drop(columns=['loan_status'])
output_data = data['loan_status']
```

```python
input_data
```

```python
output_data
```

```python
x_train,x_test,y_train,y_test=train_test_split(input_data,output_data,test_size=0.2)
```

```python
# Now data is test data set: 20% and training data set: 80%
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Now the data set is splitted into training dataset and testing dataset
```

```python
# In dataset we can see that there are large number os amounts like crores, lakhs so there will be some outliers
```

```python
# Now to overcome this problem we will do the scaling of the data which will improve the performance of the machine learning model with standard scalar this scalar function will be available in sklearn
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
# Training of the input dataset is competed
```

```python
# Now we will create the model and for that we need the logistic regression which will be imported from the sklearn
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_scaled, y_train)
```

```python
# Our model is created now let's we will do the predictions and now we will evaluate the model score with the test dataset
```

```python
model.score(x_test_scaled, y_test)
```

```python
# The accuracy of our model is more than 91%
```

```python
# Now we will predict some values
```

```python
pred_data = pd.DataFrame([['2','1','0','9600000','29900000','12','778','50700000']], columns=['no_of_dependents','education','self_employed','income_annum','loan_amount','loan_term','cibil_score','Assets'])
```

```python
pred_data=scaler.transform(pred_data)
```

```python
model.predict(pred_data)
```
